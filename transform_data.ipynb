Importamos la funcion get_coin_data y librerias necesarias

```python
from get_data import get_coin_data
import pandas as pd
```

Obtenemos el data set de las 10 monedas con  capitalizacion de mercado grande para un lapso de 4 años y guardamos la informacion en csv para su posterior uso.

```python
#Bitcoin
btc = get_coin_data('bitcoin', 1460)
#Ethereum
eth = get_coin_data('ethereum', 1460)
#Cardano
ada = get_coin_data('cardano', 1460)
#Binance Coin
bnb = get_coin_data('binancecoin', 1460)
#Uniswap
uni = get_coin_data('uniswap', 1460)
#Ripple
xrp = get_coin_data('ripple', 1460)
#Solana
sol = get_coin_data('solana', 1460)
#Polkadot
dot = get_coin_data('polkadot', 1460)
#Dogecoin
doge = get_coin_data('dogecoin', 1460)
#Litecoin
ltc = get_coin_data('litecoin', 1460)
```

Concatenamos los dataframes mediante llaves para luego poder identificar las monedas

```python
df = pd.concat([btc, eth, ada, doge, bnb , ltc, xrp, dot , sol , uni], axis=0, keys=['btc', 'eth', 'ada', 'doge', 'bnb', 'ltc', 'xrp', 'dot', 'sol', 'uni'])
```

Resultado

```python
df
```

Retoques

```python
# Redondeo a 2 decimales de price, volume , market_cap, price_change_24h, price_change_7d, price_change_30d, market_cap_change_24h, market_cap_change_7d, market_cap_change_30d, volume_change_24h, volume_change_7d, volume_change_30d
    df['price'] = round(df['price'], 2)
    df['volume'] = round(df['volume'], 2)
    df['market_cap'] = round(df['market_cap'], 2)
    df['price_change_24h'] = round(df['price_change_24h'], 2)
    df['price_change_7d'] = round(df['price_change_7d'], 2)
    df['price_change_30d'] = round(df['price_change_30d'], 2)
    df['market_cap_change_24h'] = round(df['market_cap_change_24h'], 2)
    df['market_cap_change_7d'] = round(df['market_cap_change_7d'], 2)
    df['market_cap_change_30d'] = round(df['market_cap_change_30d'], 2)
    df['volume_change_24h'] = round(df['volume_change_24h'], 2)
    df['volume_change_7d'] = round(df['volume_change_7d'], 2)
    df['volume_change_30d'] = round(df['volume_change_30d'], 2)
    # Cambio a float la columna volume y market_cap
    df['volume'] = df['volume'].astype(float)
    df['market_cap'] = df['market_cap'].astype(float)
    # Cambio el formato cientifico de volume y market_cap
    df['volume'] = df['volume'].apply(lambda x: '{:.0f}'.format(x))
    df['market_cap'] = df['market_cap'].apply(lambda x: '{:.0f}'.format(x))
```

Resultado
