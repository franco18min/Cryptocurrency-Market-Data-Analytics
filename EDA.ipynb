# Evolucion porcentual de los precios en comparacion con BTC

```python




# Crea una lista con las monedas
coins = ['btc', 'eth', 'ada', 'doge', 'bnb', 'ltc', 'xrp', 'dot', 'sol', 'uni']

# Crea una figura y un eje
fig, ax = plt.subplots()


# Itera sobre la lista de monedas
for coin in coins:
    # Selecciona los datos de precio para la moneda actual
    price_data = df.loc[coin]['price']
    
    # Grafica los datos de precio para la moneda actual
    ax.plot(price_data, label=coin.upper())

#Marca de años en la grafica
plt.xticks([0, 366, 731, 1096, 1461], [str(df['timestamp'][0]), str(df['timestamp'][366]), str(df['timestamp'][731]), str(df['timestamp'][1096]), str(df['timestamp'][1460])])

# Tamaño del grafico
fig.set_size_inches(18.5, 10.5)


ax.set_ylabel('Precio(usd)')
ax.set_xlabel('Días')

# Agrega una leyenda
ax.legend()
# Muestra el gráfico
plt.show()


```

Si hacemos una comparacion no puedo distinguir el precio de las demas monedas ya que bitcoin tiene un valor mucho mas alto que las demas , lo que si puedo observar , es como fue cambiando la tendencia de bitcoin durante los años y como esto , por ejemplo en ethereum afecto en su precio , tiene sentido ya que si el bitcoin esta en una tendencia bajista , tambien genera desconfianza sobre las demas monedas haciendo que el inversor tenga que vender por miedo , no seria un dato demasiado determinante pero si algo a tener en cuenta ya que el precio de las criptomonedas depende tambien de diferentes variables .

Pued observar como a casi finales del 2020 bitcoin tuvo un aumento bastante significativo debido a  la crisis generada por la pandemia, los bancos centrales intervinieron en los mercados financieros mediante la implementación de tipos de interés cero o negativos, la compra de deuda soberana y la impresión de moneda. Los inversores temían que esta política monetaria llevara a las economías a un escenario de inflación y devaluación de las divisas. Por eso, los inversores comenzaron a invertir en Bitcoin como lo hacen en otros activos refugio similares, como el oro.

Pero tambien nn 2021, el precio de Bitcoin experimentó una caída significativa. Una de las razones de esta caída fue una serie de eventos negativos y catalizadores, desde el CEO de Tesla, Elon Musk, hasta una nueva ronda de regulaciones por parte del gobierno chino, que afectaron a un sector de activos que se ha caracterizado por su extrema volatilidad desde su creación
