# Evolucion porcentual de los precios en comparacion con BTC

```python




# Crea una lista con las monedas
coins = ['btc', 'eth', 'ada', 'doge', 'bnb', 'ltc', 'xrp', 'dot', 'sol', 'uni']

# Crea una figura y un eje
fig, ax = plt.subplots()


# Itera sobre la lista de monedas
for coin in coins:
    # Selecciona los datos de precio para la moneda actual
    price_data = df.loc[coin]['price']
    
    # Grafica los datos de precio para la moneda actual
    ax.plot(price_data, label=coin.upper())

#Marca de aÃ±os en la grafica
plt.xticks([0, 366, 731, 1096, 1461], [str(df['timestamp'][0]), str(df['timestamp'][366]), str(df['timestamp'][731]), str(df['timestamp'][1096]), str(df['timestamp'][1460])])

# TamaÃ±o del grafico
fig.set_size_inches(18.5, 10.5)


ax.set_ylabel('Precio(usd)')
ax.set_xlabel('DÃ­as')

# Agrega una leyenda
ax.legend()
# Muestra el grÃ¡fico
plt.show()


```

Si hacemos una comparacion no puedo distinguir el precio de las demas monedas ya que bitcoin tiene un valor mucho mas alto que las demas , lo que si puedo observar , es como fue cambiando la tendencia de bitcoin durante los aÃ±os y como esto , por ejemplo en ethereum afecto en su precio , tiene sentido ya que si el bitcoin esta en una tendencia bajista , tambien genera desconfianza sobre las demas monedas haciendo que el inversor tenga que vender por miedo , no seria un dato demasiado determinante pero si algo a tener en cuenta ya que el precio de las criptomonedas depende tambien de diferentes variables .
