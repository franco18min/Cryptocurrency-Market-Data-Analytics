# Análisis Exploratorio de Datos (Supabase Version)
Este notebook demuestra cómo cargar los datos procesados directamente desde la base de datos en la nube.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

# Agregar directorio raíz al path para importar módulos de src
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from src.db.connection import get_engine

# Conectar a Supabase
engine = get_engine()

# Cargar datos
query = "SELECT * FROM cryptocurrency_prices ORDER BY price_timestamp DESC"
df = pd.read_sql(query, engine)

# Ajustar tipos de fecha
df['price_timestamp'] = pd.to_datetime(df['price_timestamp'])

print(f"Datos cargados: {df.shape}")
df.head()
```

```python
# Ejemplo: Graficar precio de Bitcoin
btc = df[df['coin'] == 'bitcoin'].sort_values('price_timestamp')

plt.figure(figsize=(12, 6))
plt.plot(btc['price_timestamp'], btc['price'], label='Precio BTC (USD)')
plt.title('Tendencia de Precio - Bitcoin')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend()
plt.grid(True)
plt.show()
```
